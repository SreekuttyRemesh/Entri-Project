5.Find if there is any correlation between age and salary , represent it visually.

```python
t=data[['Age','Salary']]
t
```

```python
Correlation=data['Age'].corr(data['Salary'])
Correlation
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x='Age', y='Salary', data=data, s=100, color='green', edgecolor='black', alpha=0.8,)
plt.title('Correlation between of Age vs Salary', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Salary', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.)
sns.regplot(x='Age', y='Salary', data=data, scatter=False, color='red')
plt.show()
```

The scatter plot with the regression line and correlation value annotation provides a visual representation of this weak positive relationship. 
The value of 0.16 indicates a weak correlation.Being positive, it indicates that as age increases, salary tends to increase as well. However, the correlation is weak, so the increase is not very strong or consistent.
